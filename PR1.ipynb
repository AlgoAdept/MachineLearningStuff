```python
#Problem Statement : Through the diagnosis test I predicted 100 reports as COVID positive ,
#but only 45 out of those were actually positive .
#Total 50 people in my sample were actually COVID positive .I have total 500 samples .
#Create confusion matrix based on above data and find 
# I. Accuracy 
#II. Precision
#III. Recall
#IV. F-1 Score
```

```python
TP=45
TN=395
FP=55
FN=5

Accuracy=(TP+TN)/(TP+TN+FP+FN)
Precision = TP/(TP+FP)
Recall = TP/(TP+FN)
Error = 1 - Accuracy
```

```python
import numpy as np
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Confusion Matrix values
TP = 45
FN = 5
FP = 55
TN = 450

# Confusion Matrix
cm = np.array([[TN, FP], [FN, TP]])

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Negative', 'Positive'])
disp.plot(cmap=plt.cm.Blues)
plt.show()

# Print confusion matrix
print("Confusion Matrix:")
print(cm)
```
